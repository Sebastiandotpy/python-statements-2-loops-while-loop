```python
def print_numbers_without_zero():
    """
    Task 1: Print numbers from -5 to 5, omitting 0.
    """
    number = -5
    while number <= 5:
        if number != 0:
            print(number, end=' ')
        number += 1
    print()

def convert_case_of_letters():
    """
    Task 1: Print numbers from -5 to 5, omitting 0.
    """
    text = input("Enter a string: ")
    converted_text = ''
    index = 0
    while index < len(text):
        char = text[index]
        if char.islower():
            converted_text += char.upper()
        elif char.isupper():
            converted_text += char.lower()
        else:
            converted_text += char
        index += 1
    print("Converted text:", converted_text)

def calculate_average():
    """
    Task 3: Calculate the average of entered integers.
    """
    print("Input integers to calculate their average.")
    print("Input 0 to exit.")
    total = 0
    count = 0
    number = None
    while number != 0:
        try:
            number = int(input("Enter an integer: "))
            if number != 0:
                total += number
                count += 1
        except ValueError:
            print("Invalid input. Please enter an integer" )

    if count > 0:
        average = total / count
        print("Average of the entered numbers:", average)
    else:
        print("No numbers entered.")

def find_character_indexes():
    """
    Task 4: Find the indexes of a character in a text.
    """                    
    text = """She came to me one morning
    One lonely Sunday morning
    Her long hair flowing in the mid-winter wind
    I know not how she found me
    For in darkness I was walking
    And destruction lay around me from a fight I could not win"""

    print("Text:\n")
    print(text)
    print()

    character = input("Enter a character to find: ")
    indexes = []
    index = 0
    while index < len(text):
        if text[index] == character:
            indexes.append(index)
        index += 1

    if len(indexes) > 0:
        print(f"\nCharacter '{character}' found at indexes:")
        for idx in indexes:
            print(idx)
    else:
        print(f"\nCharacter '{character}' not found in the text.")


def find_divisible_numbers():
    """
    Task 5: Find numbers divisible by a given divisor.
    """
    print("Enter starting and ending integers and a divisor.")
    print("Input 0 as the divisor to exit.")

    while True:
        start = int(input("Type starting integer: "))
        end = int(input("Type ending integer: "))
        divisor = int(input("Type divisor: "))

        if divisor == 0:
            print("Exiting...")
            break
        
        print()
        for number in range(start, end + 1):                 
            if number % divisor == 0:
                print(number, "is divisible by", divisor)
        print()

def run_tasks():
    print("Welcome! Select a task by entering the task number from 1 to 5.")
    print("Enter 'q' to quit.")

    while True:
        task = input("Enter task number (1-5): ")
        
        if task == 'q':
            print("Exiting...")
            break
        
        if task not in ['1', '2', '3', '4', '5']:
            print("Invalid task number. Please try again.")
            continue
        
        task_number = int(task)

        if task_number == 1:
            print("Executing Task 1:")
            print_numbers_without_zero()
        elif task_number == 2:
            print("Executing Task 2:")
            convert_case_of_letters()
        elif task_number == 3:
            print("Executing Task 3:")
            calculate_average()
        elif task_number == 4:
            print("Executing Task 4:")
            find_character_indexes()
        elif task_number == 5:
            print("Executing Task 5:")
            find_divisible_numbers()
        else:
            print("Invalid task number. Please try again.")

# Run the tasks
run_tasks()
```
